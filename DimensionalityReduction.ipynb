```python
import umap.umap_ as ump
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from gensim.downloader import load

from google.colab import drive
drive.mount('/content/drive')
```

```python
def outputGraph(df, embedding):
    fig = plt.figure(figsize=(10, 10))
    fig.suptitle(df.title)

    plt.scatter(embedding[:, 0], embedding[:, 1],
               cmap="plasma", s=2)
```

```python
def UMAPAlgorithm(df):
    umap = ump.UMAP(random_state=20)

    if df.column:
      data = df.data[:20000].drop(df.column, axis=1)
    else:
      data =  df.data[:20000]

    embedding = umap.fit_transform(data)
    outputGraph(df, embedding)
```

```python
def TSNEAlgorithm(df):
    tsne = TSNE(n_components=2, random_state=20)
    
    if df.column:
       data = df.data[:20000].drop(df.column, axis=1)
    else:
      data =  df.data[:20000]

    embedding = tsne.fit_transform(data)
    outputGraph(df, embedding)
```

```python
def PCAAlgorithm(df):
    pca = PCA(n_components=2, random_state=20)

    if df.column:
      data = df.data[:20000].drop(df.column, axis=1)
    else:
      data =  df.data[:20000]

    embedding = pca.fit_transform(data)
    outputGraph(df, embedding)
```

```python
class DataFrame:
    def __init__(self, title, data, column = None):
        self.title = title
        self.data = data
        self.column = column
```

```python
mnist = pd.read_csv(
        'https://www.openml.org/data/get_csv/52667/mnist_784.csv')
```

```python
PCAAlgorithm(DataFrame('MNIST, PCA', mnist, 'class'))
```

```python
TSNEAlgorithm(DataFrame('MNIST, tSNE', mnist, 'class'))
```

```python
UMAPAlgorithm(DataFrame('MNIST, UMAP', mnist, 'class'))
```
